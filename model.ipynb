```python
  class LinearRegression:
        X = []
        Y = []
        theta_0 = 0
        theta_1 = 0
        iterations = 0
        
        def _init_(self):
            print("beginning of the code")
    
        def addSample(self,x,y):
            self.X.append(x)
            self.Y.append(y)
            return 
        
        def getHypothesis(self,x):
            Hypothesis =  self.getTheta0() + self.getTheta_1() *x  
            return Hypothesis
        
        def getTheta0(self):
            return self.theta_0
        
        def getTheta_1(self):
            return self.theta_1
            
        def _str_(self):
            string1 = str(self.getTheta0())+ "+" +str(self.getTheta_1())       
            return string1
        
        def getSample(self):
            return self.X
        
        def getValues(self):
            return self.Y
         
        def getIteration(self):
            return self.iterations
                            
        def getLoss(self):
            loss = 0
            for i in range(len(self.X)):
                x = self.X[i]
                y = self.Y[i]
                y_pred = self.getHypothesis(x)
                y_diff = y-y_pred
                y_diff_sqred = y_diff**2
                loss = loss+y_diff_sqred
            
            mse = loss/len(self.X)
            
            return mse   
        
        def fit(self,alpha,numOfSteps):
                   
            current_hypothesis = self._str_()
            cost = self.getLoss()      
            for epoch in range(0,numOfSteps):
                sum1 = 0
                sum2 = 0
                theta_0 = self.getTheta0()
                theta_1 = self.getTheta_1() 
                for i in range(len(self.X)):
                    x = self.X[i]
                    y = self.Y[i]
                    y_pred = self.getHypothesis(x)
                    y_actual = self.Y[i]
                    sum1 = sum1+(y_actual-y_pred)
                    sum2 = sum2+(y_actual-y_pred)*x
               
                theta_0 = theta_0 - ((alpha)*(2/len(self.X))*(sum1))
                theta_1 = theta_1 - ((alpha)*(2/len(self.Y))*(sum2))
                self.theta_0 = theta_0
                self.theta_1 = theta_1
                
            self.getIteration()
                 
            return current_hypothesis,cost        
                    
```
